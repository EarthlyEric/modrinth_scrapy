```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams

data = pd.read_csv('modrinth_mods_data.csv')

ALL_LOADERS = ["Fabric", "Forge", "NeoForge", "Quilt", "LiteLoader", "Risugami's ModLoader", "Rift"]

loader_counts = data['loaders'].value_counts()
filtered_loader_counts = loader_counts[loader_counts.index.isin(ALL_LOADERS)]

rcParams['figure.figsize'] = 8, 8
rcParams['font.size'] = 10

plt.pie(filtered_loader_counts, labels=filtered_loader_counts.index, autopct='%1.1f%%', startangle=140, pctdistance=0.85, labeldistance=1.1)
plt.title('Distribution of Loaders')
plt.tight_layout()
plt.show()

for loader, count in filtered_loader_counts.items():
    percentage = (count / filtered_loader_counts.sum()) * 100
    print(f"{loader}: {percentage:.2f}%")
```
