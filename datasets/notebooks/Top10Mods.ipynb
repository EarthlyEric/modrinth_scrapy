```python
import pandas as pd
import matplotlib.pyplot as plt
from lib import parse_short

data = pd.read_csv('modrinth_mods_data.csv')
mask = ~data['categories'].str.contains('Optimization|Utility|Library', case=False, na=False) & data['categories'].notna()
filtered = data[mask]

filtered['downloads_num'] = filtered['downloads'].apply(parse_short)
top_10 = filtered.sort_values('downloads_num', ascending=False).head(10)
print(top_10[['name', 'author', 'categories', 'downloads']])

plt.figure(figsize=(10,6))
plt.barh(top_10['name'][::-1], top_10['downloads_num'][::-1], color='skyblue')
plt.xlabel('Downloads(millions)')
plt.title('The 10 Most Downloaded Mods on Modrinth(Without Optimization/Utility/Library)')
plt.tight_layout()
plt.show()
```
