```python
import pandas as pd 
import matplotlib.pyplot as plt
from lib import parse_short

data = pd.read_csv('modrinth_mods_data.csv')
data['downloads_num'] = data['downloads'].apply(parse_short)
top_authors = data.groupby('author')['downloads_num'].sum().nlargest(10)

plt.figure(figsize=(10,6))
plt.barh(top_authors.index[::-1], (top_authors.values/1e6)[::-1], color='skyblue')
plt.xlabel('Downloads (millions)')
plt.title('Top 10 Authors by Downloads')
plt.tight_layout()
plt.show()
```
